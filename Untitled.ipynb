### Herramientas utilizadas en la topografia

Para aplicar la topografia se necesitan 3 cosas :
1. <u> Estacion total</u> 
2. <u> Tripie</u> 
3. <u> Baliza</u> 

Con estas herramientas se puede aplicar la topografia.

**Estacion total** 

La <u>estacion total</u> es la encargada de el software , es la encargada de medir la inclinacion , distancias o desniveles , relieves o demas aspectos sobre el terreno ; La <u>estacion total</u> por asi decirlo es una computadora encargada de los calculos , la <u>estacion total</u> necesita del <u>tripie</u> y de la <u>Baliza</u> para su funcionaminto.

Con la <u>estacion total</u> podemos llevar un terreno del plano fisico a un programa de software llamado Civil 3D , con el cual podemos realizar planos sobre el terreno o un analisis sobre el mismo.

**Ecuacion utilizada en topografia**

Una de las ecucaiones utilizadas en topografia es *la ecuacion para determinar una distacia horizontal* entre 2 puntos.

$$ D = \sqrt{(X2 - X1)^2 + (Y2 - Y1)^2} $$

Esta ecuacion se usa mucho para situaciones en las cuales no tengamos acceso a las diferentes herramientas de topografia.
