3. Crea una funci√≥n que cree una gr√°fica y dibuje 'k' l√≠neas de $$ f(x)= sin(ax) $$La funci√≥n recibe:
x1: valor inicial de 'x'
x2: valor final de 'x'
n: cantidad de puntos del vector 'x'
a: valor inicial de image.png
b: valor final de image.png
k: cantidad de l√≠neas
Usar el la variable opcional de "label" y el comando de legend para saber qu√© l√≠nea pertenece a qu√© image.png.

```python
import matplotlib.pyplot as plt
import sympy as sp 
import numpy as np
x1=0
x2=2*np.pi
n=1000
x=np.linspace(x1,x2,n)
al = np.linspace(1,3,3)
for alpha in al:
    y = np.sin(alpha * x)
    plt.plot(x, y)
plt.show()
```
